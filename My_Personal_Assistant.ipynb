```python

def classify_questions(ques):
  goahead_with_websearch:False
  device_lst=[" alarm","reminder"," timer","call"," message",]
  personal_lst=["who are you","who created you"]

  device = False
  for i in device_lst:
   if i in ques:
    device= True

  if device:
   print("This is a Device related task!!! ")

  personal_question= False
  for i in personal_lst:
    if i in ques.lower():
      personal_question=True

  if personal_question:
    print("Hey i am your virtual aissistance created by Arry")



  if device:
      goahead_with_websearch=False
  elif personal_question:
      goahead_with_websearch= False
  else:
        goahead_with_websearch=True

  return goahead_with_websearch
```

```python
def ask_gemini(ques):
  modified_prompt = "hey give me answer to this question" + ques + "in maximum 56 words"
  response = model.generate_content(modified_prompt)
  return response.text , Markdown(response.text)
```

```python
def speak(answer):
 tts = gTTS(answer)
 tts.save('audio.mp3')
 display(Audio('audio.mp3', autoplay=True))
```

```python
have_any_other_questions = "yes"
name = ""

greetings_dict = {
    "hi": "Hi there! How can I assist you today?",
    "hello": "Hello! It's nice to hear from you. What can I do for you?",
    "hey": "Hey! What's up? How can I help?",
    "jarvis your here?" : " At your service sir!",
    "wakey wakey": "I am awake! How can I assist you today?",
    "hello jarvis": "Allow me to introduce myself. I am J.A.R.V.I.S., your virtual Artificial Intelligence. I am here to assist you with a variety of tasks as best I can, 24 hours a day, 7 days a week. Importing all preferences from home interface, systems are now fully operational."
}

while have_any_other_questions.lower() == "yes":
    if name == "":
        name = input("What is your name?-")

    def ask_questions(name):
        question = input(f"{name}, How can i help you? ")
        return question

    q = ask_questions(name)

    if q.lower() in greetings_dict:
        response = greetings_dict[q.lower()]
        speak(response)  # Speak the greeting
        display(Javascript('IPython.notebook.execute_cells_below()'))  # Refresh display
        print()  # Add a newline for visual clarity
        have_any_other_questions = input("Do you have any other questions? (yes/no): ")

    else:
        go_ahead = classify_questions(q)
        answer=""

        if go_ahead == True:
            answer_text, answer_markdown = ask_gemini(q)
            speak(answer_text)
            display(answer_markdown)
            display(Javascript('IPython.notebook.execute_cells_below()'))  # Refresh display
            print()  # Add a newline for visual clarity
            have_any_other_questions = input("Do you have any other questions? (yes/no): ")
```
